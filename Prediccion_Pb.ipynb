```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
X = dataset.drop(['Pbp'],axis = 1)
y = dataset['Pbp']

# Assuming you have your data in X and y
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.05, random_state=42)

# Further split training-validation set into training and validation
validation_split = 0.4  # Adjust this split as needed (e.g., 0.15)
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val,
                                                test_size=validation_split, random_state=42)
```

```python
# Print the number of data points in each split
print("Training data (X_train, y_train):", X_train.shape[0], y_train.shape[0])
print("Validation data (X_val, y_val):", X_val.shape[0], y_val.shape[0])
print("Test data (X_test, y_test):", X_test.shape[0], y_test.shape[0])
```

```python
#Crear el escalador
scaler = MinMaxScaler()

# Ajustar el escalador con los datos de entrenamiento
X_train_scaled = scaler.fit_transform(X_train)

# Transformar los datos de validación y prueba usando el mismo escalador
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)
# Convertir 'y' a array unidimensional si es necesario
y_train = y_train.values.reshape(-1, 1)
y_val = y_val.values.reshape(-1, 1)
y_test = y_test.values.reshape(-1, 1)
```
